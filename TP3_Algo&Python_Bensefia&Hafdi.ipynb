## **Question 1**

Créeation d'une fonction qui prend trois paramètres ``n`` et ``p`` et ``K``, entiers et retourne une liste de liste,
matrice aléatoire de ``n`` **lignes**, ``p`` **colonnes** et contenant des **entiers aléatoires compris** entre ``0`` et ``k``.

> *(Sans utilisuer **numpy** ici mais la bibliothèque **random** )*



```python
# L'imoprt de bibliotheque ``Random`` seulement pour la creation de notre matrice_NPK

from IPython.utils.path import random
from typing import List

def matrice_NPK(n:int,p:int,k:int) -> List[int]:
  matrice = []
  for ligne in range(n):
    m = []
    for colonne in range(p):
      x = random.randint(0, k)
      m.append(x)
    matrice.append(m)

  print(matrice)
```

```python
# Test de notre fonction pour vérifier que les lignes sont bien indépendantes.

# Initialisation des paramteres aleatoires
parametres = []
for p in range(3) :
  p = random.randint(1, 20)
  parametres.append(p)

# Print des paramters ainsi que la matrice_NPK
print('La matrice_NPK prend : {} lignes, {} colonnes et des entiers aleatoires compris entre 0 et {}'.format(*parametres))
n,p,k = parametres
matrice_NPK(n,p,k)
```

## **Question 3**

Créeation d'une fonction qui réalise une multiplication entre deux matrices représentez sous forme de
liste de liste.

> *(Sans utilisuer **numpy**)*

```python
def multiplication(a,b):
  try:
    # Condition nécessaire pour la multiplication

      ncol_mat1 = len(a[0])
      nlig_mat2 = len(b)
      if(ncol_mat1 != nlig_mat2):
          raise ValueError("Multiplication impossible, nombre de colonnes de la 1e matrice est different du nombre de lignes de la 2e matrice")

      #matrice résultate vide
      matrice=[[0 for c in range(len(b[0]))] for l in range(len(a))]

      for i in range(len(a)):
        for j in range(len(b[0])):
          for k in range(len(b)):
            matrice[i][j] += a[i][k] * b[k][j]

      return matrice
  except ValueError:
      print("Multiplication impossible, nombre de colonnes de la 1e matrice est different du nombre de lignes de la 2e matrice")
  print(matrice)
```

# **Question 4**

> Créeation de la meme fonction en utilisant **``Numpy``**.

```python
def multiplication_np(a,b):
  resultat = np.dot(a,b)
  print(resultat)
```

# **Question 5**

>  évaluation des temps de calcul des multiplications pour des matrices aléatoires **``n*n``** avec **``n``** croissant, pour les deux méthodes et tracage sur le même graphique les deux courbes des temps de calcul.


```python
import matplotlib.pyplot as plt
import timeit

tailles = [1,2,3,4,5,10]

temps_execution_mutilplication_simple = []
temps_execution_multiplication_np = []

for taille in tailles:
    matrice = [[taille for c in range(taille)] for l in range(taille)]
    liste_a_trier = [random.randint(0, 1000) for _ in range(taille)]

    # Mesure du temps d'exécution de la multiplication simple
    temps_mutilplication_simple = timeit.timeit(lambda: multiplication(matrice,matrice))
    temps_execution_mutilplication_simple.append(temps_mutilplication_simple)

    # Mesure du temps d'exécution de la multiplication numpy
    temps_multiplication_np = timeit.timeit(lambda: multiplication_np(matrice,matrice), number=1)
    temps_execution_multiplication_np.append(temps_multiplication_np)

# Tracer le graphique
plt.plot(tailles, temps_execution_mutilplication_simple, marker='o', label='mutilplication simple')
plt.plot(tailles, temps_execution_multiplication_np, marker='o', label='multiplication avec numpy')
plt.xlabel('Taille du tableau')
plt.ylabel('Temps d\'exécution (secondes)')
plt.title('Temps d\'exécution des méthodes multiplication des matrices en fonction de leurs tailles')
plt.grid(True)
plt.legend()
plt.show()
```

# **Question 6**
L'ajout qui prend en paramètre une matrice et lui rajoute une ligne en-dessous.


> Sans utiliser **``Numpy``**.

```python
def ajout_ligne(a:list,b:list):
  try:
    # Condition nécessaire pour l'ajout
      ncol_matrice = len(a[0])
      ncol_ligne = len(b)
      if(ncol_matrice != ncol_ligne):
          raise ValueError("Concaténation impossible, nombre de colonnes de la matrice est different de celui de la ligne rajoutée")
      r = a.append(b)
      print(r)
  except ValueError:
      print("Concaténation impossible, nombre de colonnes de la matrice est different de celui de la ligne rajoutée")
```

> En utilisant **``Numpy``**.

```python
def ajout_ligne_np(a,b):
  return np.concatenate((a, b), axis=0)
```

évaluation des temps de calcul des deux méthodes et tracage sur le même graphique les deux courbes des temps de calcul :

```python
tailles = [50,100,150,200,500]

temps_execution_ajout_simple = []
temps_execution_ajout_np = []

for taille in tailles:
    matrice = [[taille for c in range(taille)] for l in range(taille)]
    ligne = [taille for k in range(taille)]
    ligne_array = np.array(ligne)
    liste_a_trier = [random.randint(0, 1000) for _ in range(taille)]

    # Mesure du temps d'exécution de la multiplication simple
    temps_ajout_simple = timeit.timeit(lambda: ajout_ligne(matrice,ligne), number=1)
    temps_execution_ajout_simple.append(temps_ajout_simple)

    # Mesure du temps d'exécution de la multiplication numpy
    temps_ajout_np = timeit.timeit(lambda: ajout_ligne_np(np.array(matrice),ligne_array.reshape(1,ligne_array.shape[0])), number=1)
    temps_execution_ajout_np.append(temps_ajout_np)

# Tracer le graphique
plt.plot(tailles, temps_execution_ajout_simple, marker='o', label='ajout simple')
plt.plot(tailles, temps_execution_ajout_np, marker='o', label='ajout avec numpy')
plt.xlabel('Taille du tableau')
plt.ylabel('Temps d\'exécution (secondes)')
plt.title('Temps d\'exécution des méthodes pour rajouter une ligne à des matrices en fonction de leurs tailles')
plt.grid(True)
plt.legend()
plt.show()
```

On peut conclure que **`Numpy`** est plus efficace en terme de calcul matriciel, lors que l'utilisation des listes est a eu des meilleurs résultats quand on l'utilise pour des fins d'écriture, sauvgarde des données ou meme pour les modifications ..
